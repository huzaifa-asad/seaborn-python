# Seaborn Visualization Tutorial

This notebook provides a step-by-step guide to data visualization using the Seaborn library, which is built on top of Matplotlib and integrates well with Pandas data structures.

You will learn how to create line plots, bar plots, and histograms, and how to customize your charts using Seaborn's powerful features.

# Seaborn Library:
   > derived from Matplotlib and Pandas libraries.

## Why Import These Libraries?
- **Seaborn**: High-level interface for drawing attractive statistical graphics.
- **Matplotlib**: Underlying plotting library; Seaborn uses it for rendering.
- **Pandas**: For data manipulation and DataFrame support, which Seaborn works well with.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Creating Sample Data
We start with simple Python lists for `x` and `y` to demonstrate basic plotting. These represent the values to be plotted on the x-axis and y-axis, respectively.

```python
x = [1, 2, 3, 4, 5,6, 7, 8]
y = [8, 7, 6, 5, 4, 3, 2, 1]
```

### Plotting with Matplotlib
Matplotlib is the foundation for most Python plotting libraries. Here, we use `plt.plot()` to create a simple line plot.

```python
# matplotlib
plt.plot(x, y)
plt.show()
```

### Plotting with Seaborn
Seaborn provides a simpler and more attractive interface for statistical plots. `sns.lineplot()` automatically adds features like confidence intervals and color palettes.

### In Seaborn lib:

```python
sns.lineplot(x=x, y=y)
```

### Using Pandas DataFrames with Seaborn
Seaborn works seamlessly with Pandas DataFrames. This allows you to plot data using column names, making code more readable and flexible.

```python
import pandas as pd 
x = [1, 2, 3, 4, 5,6, 7, 8]
y = [8, 7, 6, 5, 4, 3, 2, 1]
df = pd.DataFrame({
    'var': x,
    'var1': y
})
df
```

### Line Plot from DataFrame Columns
Here, we use column names from the DataFrame to plot. This is especially useful for larger datasets and makes code easier to maintain.

```python
sns.lineplot(x='var', y='var1', data=df)
plt.show()
```

### Example Dataset: Penguins
Seaborn includes several built-in datasets for practice. The 'penguins' dataset contains measurements for different penguin species and is commonly used for visualization demos.

```python
df1 = sns.load_dataset("penguins")
df1
```

### Grouping and Customizing Lines
- **hue**: Colors lines by a categorical variable (e.g., 'sex').
- **style**: Changes line style or marker by category.
- **palette**: Sets the color scheme.
These options help visualize differences between groups.

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm',hue='sex', data=df1)
plt.show()
```

### Using Markers for Categorical Distinction
Markers help distinguish categories visually. You can specify marker shapes for each category using the `markers` parameter.

```python
# style parameter is using for markers to specifiy markers..
sns.lineplot(x='bill_length_mm', y='flipper_length_mm',hue='sex', data=df1, palette='rocket_r', style='sex', markers=["o", "<"])
```

### Plotting a Subset of Data
Sometimes, it's useful to plot only a subset of your data for clarity or performance. Here, we use the first 20 rows of the penguins dataset.

```python
df1 = sns.load_dataset("penguins").head(20)
df1
```

### Adding Titles to Plots
Adding a title helps viewers understand what the plot represents. Use `plt.title()` to set a descriptive title for your chart.

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm',hue='sex', data=df1, palette='rocket_r',
             style='sex', markers=["o", "<"])
plt.title("Line Chart of Penguins data")
plt.show()
```

### Reloading Data for Bar Plots
We reload the full penguins dataset to demonstrate bar plots, which are useful for comparing categorical data.

```python
df1 = sns.load_dataset("penguins")
df1
```

### Bar Plots: Comparing Categories
Bar plots are ideal for comparing values across categories. The `hue` parameter allows you to split bars by another categorical variable, such as 'sex'.

```python
df1["island"]
```

### Customizing Category Order
You can control the order in which categories appear using the `order` parameter. This is useful for emphasizing certain groups or for presentation purposes.

```python
sns.barplot(x="island", y="bill_length_mm",hue="sex", data=df1)
plt.show()
```

### Customizing Hue Order
The `hue_order` parameter lets you specify the order of categories for the hue variable, ensuring consistent color mapping across plots.

```python
# Order parameter use to show bars as you want.

order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df1, hue="sex",order=order1)
plt.show()
```

### Setting Bar Colors
You can set the color of all bars using the `color` parameter. This is useful for highlighting specific data or matching a theme.

```python
# hue order 
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df1, hue="sex",order=order1, hue_order=['Female', 'Male'])
plt.show()
```

### Using Color Palettes
Seaborn offers many built-in color palettes for categorical data. The `palette` parameter lets you choose a color scheme that improves readability and aesthetics.

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df1, order=order1, color="red")
plt.show()
```

### Adjusting Color Saturation
The `saturation` parameter controls the intensity of bar colors. Lower saturation makes colors lighter, which can help with visual clarity.

```python
sns.barplot(x="island", y="bill_length_mm",data=df1, hue="sex", palette="prism")
```

# Histograms: Visualizing Distributions
Histograms show the distribution of a single variable. They are useful for understanding the spread and central tendency of your data.

```python
# Saturation parameter 
sns.barplot(x="island", y="bill_length_mm",data=df1, hue="sex", palette="prism", saturation=0.1)
```

# Histogram Plot

```python
sns.displot(df1["flipper_length_mm"])
```

### Customizing Histogram Bins and Colors
- **bins**: Define the intervals for grouping data.
- **color**: Set the color of the histogram bars.
Customizing these parameters helps highlight specific features in your data.

```python
# bins and color
sns.displot(df1["flipper_length_mm"], bins=[170, 180, 190,200, 210, 220, 230, 240], color="g")
```

### Adding Kernel Density Estimate (KDE)
KDE is a smooth curve that estimates the probability density function of a variable. Adding `kde=True` overlays this curve on the histogram for better visualization of the distribution shape.

```python
# kde ?
sns.displot(df1["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240], color="g", kde=True)
```
